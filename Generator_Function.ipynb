```python
def test_fib(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b = b , a+b
```

```python
test_fib(45)
```

```python
for i in test_fib(10):
  print(i, end=" ")
```

```python

def test_fib1():
  a,b = 0,1
  while True:
    yield a
    a,b = b, a+b
```

```python
fib = test_fib1()
```

```python
print(fib)
```

```python
for i in range(10):
  print(next(fib))
```

```python
def count_test(n):
  count = 1
  while count <= n :
    yield count
    count = count + 1
```

```python
c  = count_test(5)
```

```python
type(c)
```

```python
for i in c:
  print(i)
```
